```python
data_x = train.drop('Survived', axis=1)
data_y = train['Survived']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

import numpy as np

search = GridSearchCV(
    DecisionTreeClassifier(),
    {
        'criterion': ['gini'],
        'max_depth': np.arange(1, 15)
    },
    cv=10,
    scoring='accuracy'
)

search.fit(data_x, data_y)
print(search.best_estimator_.max_depth)
```

```python
search = GridSearchCV(
    DecisionTreeClassifier(),
    {
        'criterion': ['entropy'],
        'max_depth': np.arange(1, 15)
    },
    cv=10,
    scoring='accuracy'
)

search.fit(data_x, data_y)
print(search.best_estimator_.max_depth)
```

```python
# 6
from sklearn.metrics import accuracy_score

decision_tree = DecisionTreeClassifier(max_depth=3, criterion='entropy', min_samples_leaf=1)
decision_tree.fit(data_x, data_y)

pred_data_y = decision_tree.predict(data_x)
accuracy_score(pred_data_y, data_y)
```

```python
# 6
f = open("pred_data_survivers.txt", "w")
f.write(str(pred_data_y))
f.flush()
f.close()

pred_data_y
```

```python
# 7
from sklearn import tree

fig = plt.figure(figsize=(25,20))
tree.plot_tree(decision_tree=decision_tree, feature_names=data_x.columns, class_names=['Died', 'Survived'])
```

```python
# 8
accuracies = []
params = []

for criterion in [0, 1]:
    for max_depth in np.arange(1, 15):
        for min_samples_leaf in [1, 5, 10, 15]:
            criterion_label = 'gini' if criterion == 0 else 'entropy'

            model = DecisionTreeClassifier(criterion=criterion_label, max_depth=max_depth, min_samples_leaf=min_samples_leaf, random_state=42)
            model.fit(data_x, data_y)
            
            y_pred = model.predict(data_x)
            accuracy = accuracy_score(data_y, y_pred)            
            accuracies.append(accuracy)
            params.append((criterion, max_depth, min_samples_leaf))

accuracies = np.array(accuracies)
params = np.array(params)

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(params[:, 0], params[:, 1], params[:, 2], c=accuracies, cmap='coolwarm', s=50)
ax.set_xlabel('0 = Gini, 1 = Entropy')
ax.set_ylabel('Max Depth')
ax.set_zlabel('Min Samples Leaf')
ax.set_title('Accuracy to parameters chart')

cbar = plt.colorbar(ax.collections[0], ax=ax, orientation='horizontal')
cbar.set_label('Accuracy')

plt.show()
```

```python
# 9
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, train_size=0.75)
```

```python
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import precision_recall_fscore_support

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.3f}')

f1_sc = f1_score(y_test, y_pred)
print(f'F1-score: {f1_sc:.3f}')

precision_avg, recall_avg, _, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted')
print(f'Avg Precision: {precision_avg:.3f}')
print(f'Avg Recall: {recall_avg:.3f}')
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2, random_state=42)

# Define parameter grids to explore
criteria = [0, 1]
max_depths = np.arange(1, 10)
min_samples_leafs = [1, 5, 10, 15]

# Create lists to store results
results = []
param_combinations = []

# Iterate over parameter combinations
for criterion in criteria:
    for max_depth in max_depths:
        for min_samples_leaf in min_samples_leafs:
            criterion_label = 'gini' if criterion == 0 else 'entropy'
            # Train a Decision Tree classifier with current parameters
            model = DecisionTreeClassifier(criterion=criterion_label, max_depth=max_depth, min_samples_leaf=min_samples_leaf, random_state=42)
            model.fit(X_train, y_train)
            
            # Evaluate the model on the test set
            y_pred = model.predict(X_test)
            accuracy = accuracy_score(y_test, y_pred)
            
            # Store results and parameter combination
            results.append(accuracy)
            param_combinations.append((criterion, max_depth, min_samples_leaf))

# Convert results and parameter combinations to numpy arrays for plotting
results = np.array(results)
param_combinations = np.array(param_combinations)

# Extract individual parameter arrays
criteria_array = param_combinations[:, 0]
max_depths_array = param_combinations[:, 1]
min_samples_leafs_array = param_combinations[:, 2]

# Plotting the results on a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of parameter combinations with accuracy as the z-axis
ax.scatter(criteria_array, max_depths_array, min_samples_leafs_array, c=results, cmap='viridis', s=50)
ax.set_xlabel('Criterion')
ax.set_ylabel('Max Depth')
ax.set_zlabel('Min Samples Leaf')
ax.set_title('Influence of DecisionTreeClassifier Parameters')

# Adding a color bar to show accuracy values
cbar = plt.colorbar(ax.collections[0], ax=ax, orientation='vertical')
cbar.set_label('Accuracy')

plt.show()
```
